```python
# Import graphics by assgining it to g
import graphics as g

# Create a function where text will be. 
def text(x1, y1, win, word):
# Put in points, backgrounds, and word for the diffrent places.
    txt = g.Text(g.Point(x1, y1), word)
    txt.draw(win)

# Create a function for the rectangle.
def Rectangle(x1, y1, l, w, win):
# Put in points and add width and length to the varialble, and add the backgrounds.
    rect = g.Rectangle(g.Point(x1,y1), g.Point(x1+w, y1+l))
# Assgin color for line rectangle.
    rect.setOutline(g.color_rgb(0,225,225))
# Assgin the thickness of the triangle.
    rect.setWidth(2)
# Fill a color for the rectangle.
    rect.setFill(g.color_rgb(225,100,50))
    rect.draw(win)

# Create a main function where it call all the other function.
def main():
# Assgin a title for the graphics "sudoku solver"
    win = g.GraphWin("Sudoku Solver", 500,500)
# Assgine a color for the background 
    win.setBackground(g.color_rgb(100,150,150)) #customizing colours no more than 225

# Create a for loop to repeat the first rectangle 9 times   
    for i in range(9):
# Create another for loop and adjust the points 
        for j in range(9):
            Rectangle(160+(i*20), 160+(j*20), 20, 20, win)
            print()
# Call the rectangle function for the easy box.           
    Rectangle(80, 90, 40, 50, win) 
# Call the text function and assgin a word.
    text(105, 110, win, "easy")
# Call the rectangle function for the medium box. 
    Rectangle(230, 90, 40, 50, win)  
# Call the text function and assgin a word.
    text(255, 110, win, "medium")
# Call the rectangle function for the hard box. 
    Rectangle(360, 90, 40, 50, win)
# Call the text function and assgin a word.
    text(385, 110, win, "hard")

# Call the rectangle function for the start game box. 
    Rectangle(140, 20, 40, 70, win)
# Call the text function and assgin a word.
    text(175, 40, win, "start game")
# Call the rectangle function for the timer box. 
    Rectangle(300, 20, 40, 50, win)
# Call the text function and assgin a word.
    text(325, 40, win, "timer")

#  Get mouse 
    win.getMouse()
    win.close()
# Call the main function   
main()
```
