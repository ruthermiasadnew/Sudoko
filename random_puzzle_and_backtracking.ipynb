```python
##########      FUNCTIONS USED IN MAIN      ##########
def check_line(grid, location, value):
    '''This function accepts two inputs of a grid of lists of lists (4x4) and a string of the location we
    want to check. which can be like '00' or '32', where the first location is the first of the string    
    '''
    error = []      #An empty list that will contain all the errors on that location
    
    #mover = [[1,2,3],[10,20,30]]
    #for i in range(4):
    #    for location in mover:         #Loop through the horizontal and vertial potion of the 00 position
    #        for value in location:             
    #            print(starter+value)   #These are the location we need to check for the 00 position
    #    mover[0][i] *= -1
    #    print(mover, starter)
    

    x = int(location[1])    #Save the string into x and y values
    y = int(location[0])

         #check if the grid is different from 0, so it will only check filled grids
    #horizonal check
    for hor in range(9):    #loop through the 0 to 3
        if hor != x:        #Avoid checking the current position so if it is in '00' don't want to check '00' again
            if grid[y][hor] == value:          #Check if the location passed is equal to every value in the horizontal
                error.append(str(y)+str(hor))       #If so it is an error  

    #vertical check
    for ver in range(9):    #loop through 0 to 3
        if ver != y:        #Avoid checking the current position so if it is in '00' don't want to check '00' again (in the y axis)
            if grid[ver][x] == value:      #Check if the location passed is equal for very value in vertical
                error.append(str(ver)+str(x))   #Add to error

    #Check per box
    if x//3 == 0 and y//3 == 0:         #Check for box 1, by divide the values of the x and y values we can guess where in the box it is found
        for hor in range(3):        #For box 1 only values 0 and 1 are needed
            for ver in range(3):
                if hor != x and ver != y:       #Check if it is equal to the passed value
                    if grid[ver][hor] == value:        #Check if the passed location's value is equal to all the value in the box
                        error.append(str(ver)+str(hor))     #If so add it into error

    elif x//3 == 1 and y//3 == 0:      #Check for box 2
        for hor in range(3,6):
            for ver in range(3):    #For box 2: 0,1 for the vertical and horizional value 2,3 is what needed and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))

    elif x//3 == 2 and y//3 == 0:      #Check for box 3
        for hor in range(6,9):
            for ver in range(3):    #For box 3: 0,1 for the vertical and horizional value 2,3 is what needed and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))

    elif x//3 == 0 and y//3 == 1:      #Check for box 4
        for hor in range(3):
            for ver in range(3,6):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))
    elif x//3 == 1 and y//3 == 1:
        for hor in range(3,6):      #Check for box 5
            for ver in range(3,6):  #For box 4: 2,3 is what needed for hor and vert and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))

    elif x//3 == 1 and y//3 == 1:
        for hor in range(6,9):      #Check for box 6
            for ver in range(3,6):  #For box 4: 2,3 is what needed for hor and vert and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))
        
    elif x//3 == 0 and y//3 == 2:      #Check for box 4
        for hor in range(3):
            for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))

    elif x//3 == 1 and y//3 == 2:      #Check for box 4
        for hor in range(3,6):
            for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))

    elif x//3 == 2 and y//3 == 2:      #Check for box 4
        for hor in range(6,9):
            for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                if hor != x and ver != y:
                    if grid[ver][hor] == value:
                        error.append(str(ver)+str(hor))
        
    if len(error)>0:
        return 1
    else:
        return 0

#####
    
def zero_finder(grid):
    '''finds every empty space on a grid, adds it to a list'''
    zero_spots = []
    for a in range(81):
        row = a//9
        col=a%9
        if grid[row][col]==0:
            zero_spots.append(str(row)+str(col))
    return zero_spots

#####

def full(grid):
    '''receives a grid and returns 0 if there are still empty spaces'''
    for row in range(9):
        for col in range(9):
            if grid[row][col]==0:
                return 0
    return 1

#####

def create_puzzle(puzzle):
    '''receives a string input of 81 numerical characters and returns a 9x9 list of lists'''
    multi_list = []
    #for every 9th number, starting from 0 and going to 72
    for a in range(0,81,9):
        #reset single list
        single_list = []
        #add each grouping of nine (0-9, 10-18, etc.) to the single list
        for b in puzzle[a:a+9]:
            single_list.append(int(b))
        #add the single list to the big list
        multi_list.append(single_list)
    return multi_list

#####

def pull_puzzle(difficulty):
    '''receives input as easy, medium, or hard then returns a random puzzle based off that'''
    #create the file name by combining the input string and the .txt extension
    file_name = difficulty + '.txt'
    #using file name, choose which file to open and create an empty list for each line
    file = open(file_name)
    #add each line to a list
    file_list = file.readlines()
    #return a random puzzle from the list, stripping the /n off
    return file_list[random.randrange(0,len(file_list))].strip()
  
#####
    
def backtracking(puzzle):
    #creates a list of coordinates for all initially empty tiles
    empty_spots = zero_finder(puzzle)
    #current defines what coordinate is being used
    current = 0
    error = 1
    while error != 0:    
        #set current row and column values based off current and the empty spot list
        row = int(empty_spots[current][0])
        col = int(empty_spots[current][1])
        
        #print statements used for debugging
        #displays current, row, col, the current empty spot and the puzzle so far
        #print('current', current)
        #print('coordinate', str(row)+str(col))
        #print('pre', empty_spots[current], puzzle)
        
        #increment current tile by one then set value equal to the current tile
        puzzle[row][col] += 1
        value = puzzle[row][col]
        
        #print statements used for debugging
        #displays current empty tile and puzzle post tile increment
        #print('post', empty_spots[current], puzzle)
        
        #if the current value is greater than the puzzle allows, reset the tile and step back
        if value < 10:
            #if the current tile has no errors, increment current
            if check_line(puzzle, empty_spots[current], value) == 0:
                #print('no errors found. next tile')

                #print statements used for debugging
                #displays current before increment
                #print('pre current', current)

                #increment current
                current += 1

                #print statements used for debugging
                #displays current after increment
                #print('post current', current)

        else:
            #print('value greater than four. attempting reset')
            
            #print statements used for debugging
            #displays current and tile value pre-adjustment
            #print('pre current', current)
            #print('pre value', puzzle[row][col])
            
            #resets tile value to zero and decrements current in order to move to previous empty space
            puzzle[row][col] = 0
            current = current - 1
            
            #print statements used for debugging
            #displays current and tile value post-adjustment
            #print('post current', current)
            #print('post value', puzzle[row][col])
        
        #if the puzzle is full, check every tile in empty_spots to make sure there are no errors
        if full(puzzle) != 0:
            #count the amount of errors, set error = 0 initially
            error = 0
            #for every tile, add 1 for every time an error occurs
            for a in empty_spots:
                error += check_line(puzzle, a, puzzle[int(a[0])][int(a[1])])
        
        #print('END OF WHILE LOOP')
    print("SOLVED PUZZLE")
    for i in range(9):
        print(puzzle[i])

##########     END OF FUNCTIONS IN MAIN     #########
import random

def main():
    '''receives desired difficulty, chooses a random puzzle, solves the puzzle using backtracking'''
    #choose a difficulty
    #THIS WILL BE REPLACED WITH THE BUTTON PUSH ON THE GRAPHICS
    #MAKE SURE IT RETURNS A STRING THAT SAYS 'EASY', 'MEDIUM', OR 'HARD'
    difficulty = input("Enter difficulty (easy, medium, hard):")
    #the raw 81 character puzzle is chosen
    raw_puzzle = pull_puzzle(difficulty)
    #then converted into a list of lists
    grid_puzzle = create_puzzle(raw_puzzle)
    #then sent to the backtracking function and solved
    backtracking(grid_puzzle)
    
main()
```
