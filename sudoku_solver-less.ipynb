```python
#GRAPHICS BY REDIET AND RUTH
#STOCHASTIC BY ABEMELECH AND ALEC KRAY
#BACKTRACKING BY BEREKET AND ROBYN

#################### Functions used in BACKTRACKING ####################
#################### Abemelech's Code ####################
class Backtracking:
    def check_line(grid, location, value): 
        '''This function accepts two inputs of a grid of lists of lists (4x4) and a string of the location we
        want to check. which can be like '00' or '32', where the first location is the first of the string   
        '''
        error = []      #An empty list that will contain all the errors on that location
        
        #mover = [[1,2,3],[10,20,30]]
        #for i in range(4):
        #    for location in mover:         #Loop through the horizontal and vertial potion of the 00 position
        #        for value in location:             
        #            print(starter+value)   #These are the location we need to check for the 00 position
        #    mover[0][i] *= -1
        #    print(mover, starter)
        

        x = int(location[1])    #Save the string into x and y values
        y = int(location[0])

            #check if the grid is different from 0, so it will only check filled grids
        #horizonal check
        for hor in range(9):    #loop through the 0 to 9
            if hor != x:        #Avoid checking the current position so if it is in '00' don't want to check '00' again
                if grid[y][hor] == value:          #Check if the location passed is equal to every value in the horizontal
                    error.append(str(y)+str(hor))       #If so it is an error  

        #vertical check
        for ver in range(9):    #loop through 0 to 9
            if ver != y:        #Avoid checking the current position so if it is in '00' don't want to check '00' again (in the y axis)
                if grid[ver][x] == value:      #Check if the location passed is equal for very value in vertical
                    error.append(str(ver)+str(x))   #Add to error

        #Check per box
        if x//3 == 0 and y//3 == 0:         #Check for box 1, by divide the values of the x and y values we can guess where in the box it is found
            for hor in range(3):        #For box 1 only values 0 to 2 are needed
                for ver in range(3):
                    if hor != x and ver != y:       #Check if it is equal to the passed value
                        if grid[ver][hor] == value:        #Check if the passed location's value is equal to all the value in the box
                            error.append(str(ver)+str(hor))     #If so add it into error

        elif x//3 == 1 and y//3 == 0:      #Check for box 2
            for hor in range(3,6):
                for ver in range(3):    #For box 2: 0,1,2 for the vertical and horizional value 3,4,5 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))

        elif x//3 == 2 and y//3 == 0:      #Check for box 3
            for hor in range(6,9):
                for ver in range(3):    #For box 3: 0,1,2 for the vertical and horizional value 6,7,8 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))

        elif x//3 == 0 and y//3 == 1:      #Check for box 4
            for hor in range(3):
                for ver in range(3,6):  #For box 3: 0,1,2 for the horizonal and vertical value 3,4,5 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))
        elif x//3 == 1 and y//3 == 1:
            for hor in range(3,6):      #Check for box 5
                for ver in range(3,6):  #For box 4: 3,4,5 is what needed for hor and vert and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))

        elif x//3 == 1 and y//3 == 1:
            for hor in range(6,9):      #Check for box 6
                for ver in range(3,6):  #For box 4: 6,7,8 for the horizonal and vertical value 3,4,5 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))
            
        elif x//3 == 0 and y//3 == 2:      #Check for box 4
            for hor in range(3):
                for ver in range(6,9):  #For box 3: 0,1,2 for the horizonal and vertical value 6,7,8 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))

        elif x//3 == 1 and y//3 == 2:      #Check for box 4
            for hor in range(3,6):
                for ver in range(6,9):  #For box 3: 3,4,5 for the horizonal and vertical value 6,7,8 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))

        elif x//3 == 2 and y//3 == 2:      #Check for box 4
            for hor in range(6,9):
                for ver in range(6,9):  #For box 3: 6,7,8 for the horizonal and vertical value 6,7,8 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == value:
                            error.append(str(ver)+str(hor))  

    #################### END of Abemelech's Code ####################

    #################### Bereket's Code ####################        
        if len(error)>0:
            return 1        #Return True if error
        else:
            return 0        #Return False if error

    #####

    def zero_finder(grid):
        '''finds every empty space on a grid, adds it to a list'''
        zero_spots = []     #Empty List
        for a in range(81):     #Loop through the entire grid
            row = a//9
            col=a%9
            if grid[row][col]==0:       #Check if 0
                zero_spots.append(str(row)+str(col))    #Added to the list
        return zero_spots   #Return the zero list

    #####

    def full(grid):
        '''receives a grid and returns 0 if there are still empty spaces'''
        for row in range(9):
            for col in range(9):        #Loop through the entire grid
                if grid[row][col]==0:   #If there is  zero then return 0
                    return 0
        return 1                        #No zero return 1
    #################### END of Bereket's Code ####################
#################### END of functions in BACKTRACKING ####################

#################### Functions used in STOCHASTIC ####################
#################### Abemelech's Code ####################
import random
class Stochastic:
    def check_line(grid, location):
        '''This function accepts two inputs of a grid of lists of lists (4x4) and a string of the location we
        want to check. which can be like '00' or '32', where the first location is the first of the string    
        '''
        error = []      #An empty list that will contain all the errors on that location
        
        #mover = [[1,2,3],[10,20,30]]
        #for i in range(4):
        #    for location in mover:         #Loop through the horizontal and vertial potion of the 00 position
        #        for value in location:             
        #            print(starter+value)   #These are the location we need to check for the 00 position
        #    mover[0][i] *= -1
        #    print(mover, starter)
        

        x = int(location[1])    #Save the string into x and y values
        y = int(location[0])

        if grid[y][x] != 0:         #check if the grid is different from 0, so it will only check filled grids
            #horizonal check
            for hor in range(9):    #loop through the 0 to 3
                if hor != x:        #Avoid checking the current position so if it is in '00' don't want to check '00' again
                    if grid[y][hor] == grid[y][x]:          #Check if the location passed is equal to every value in the horizontal
                        error.append(str(y)+str(hor))       #If so it is an error  

            #vertical check
            for ver in range(9):    #loop through 0 to 3
                if ver != y:        #Avoid checking the current position so if it is in '00' don't want to check '00' again (in the y axis)
                    if grid[ver][x] == grid[y][x]:      #Check if the location passed is equal for very value in vertical
                        error.append(str(ver)+str(x))   #Add to error

            #Check per box
            if x//3 == 0 and y//3 == 0:         #Check for box 1, by divide the values of the x and y values we can guess where in the box it is found
                for hor in range(3):        #For box 1 only values 0 and 1 are needed
                    for ver in range(3):
                        if hor != x and ver != y:       #Check if it is equal to the passed value
                            if grid[ver][hor] == grid[y][x]:        #Check if the passed location's value is equal to all the value in the box
                                error.append(str(ver)+str(hor))     #If so add it into error

            elif x//3 == 1 and y//3 == 0:      #Check for box 2
                for hor in range(3,6):
                    for ver in range(3):    #For box 2: 0,1 for the vertical and horizional value 2,3 is what needed and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))

            elif x//3 == 2 and y//3 == 0:      #Check for box 3
                for hor in range(6,9):
                    for ver in range(3):    #For box 3: 0,1 for the vertical and horizional value 2,3 is what needed and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))


            elif x//3 == 0 and y//3 == 1:      #Check for box 4
                for hor in range(3):
                    for ver in range(3,6):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))
            elif x//3 == 1 and y//3 == 1:
                for hor in range(3,6):      #Check for box 5
                    for ver in range(3,6):  #For box 4: 2,3 is what needed for hor and vert and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))

            elif x//3 == 1 and y//3 == 1:
                for hor in range(6,9):      #Check for box 6
                    for ver in range(3,6):  #For box 4: 2,3 is what needed for hor and vert and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))
            
            elif x//3 == 0 and y//3 == 2:      #Check for box 4
                for hor in range(3):
                    for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))

            elif x//3 == 1 and y//3 == 2:      #Check for box 4
                for hor in range(3,6):
                    for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))

            elif x//3 == 2 and y//3 == 2:      #Check for box 4
                for hor in range(6,9):
                    for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                        if hor != x and ver != y:
                            if grid[ver][hor] == grid[y][x]:
                                error.append(str(ver)+str(hor))
            
            return error

    #print(grid)

    def count_grid(grid):
        ''' This counts the number of values
        Ex: there is 2 0's and 3 4's and so forth and so one.
        It passed the grid to check for how much of a single value is available 
        then it stored in a dictionary
        '''
        counter = {}        #Empty dictionary
        for hor in grid:    #Loop through all the values of the grid  
            for value in hor:
                if value != 0:
                    if value in counter:
                        counter[value] += 1     #Add one if the value exists
                    else:
                        counter[value] = 1      #Create one if the value doesn't exist
        return counter

    def create_missing(counter):
        '''The function creates a list of missing values that could
        later be filled. It accepts a dictionary of the current available values
        '''
        possible_value = []         #Empty list to store the values
        for key,value in counter.items():   #loop through the dictionary
            for _ in range(9-value):        #Loop through the number of times it is empty and add the value to the list
                possible_value.append(key) 
        return possible_value
    #################### END of Abemelech's Code ####################
    #################### Alec Kray's Code ####################

    def original(puzzle):
        '''Create List of original puzzle values'''
        org = []        #List to store
        for i in range(9):      #Loop through the grid to store the original puzzle location
            for j in range(9):
                if puzzle[i][j] != 0:
                    org.append(str(i)+str(j))
        return org

    def check_zero(puzzle):
        '''Loop through the puzzle and find the location of the zeros'''
        zero = []
        for i in range(9):      #Loop throug the entire puzzle
            for j in range(9):
                if puzzle[i][j] == 0:       #Check those that are zeros
                    zero.append(str(i)+str(j))          #Store the location of the zeros

        return zero

    def filler(puzzle,li):
        '''Fill the puzzle randomly based on a list'''
        for horizontal in puzzle:   #Loop through the puzzle
            for i in range(9):      
                #print(horizontal[i])
                if horizontal[i] == 0:  #Check if puzzle value is zero
                    horizontal[i] = random.choice(li)     #Fill randomly from list       
                    li.remove(horizontal[i])    #Remove the number from the list since we don't want duplicates

    def checker(puzzle,li):
        ''' Check the entire puzzle if there is an error'''
        error = []      #Empty list
        for i in range(9):      #Loop through i and j 0 to 3
            for j in range(9):
                if str(i)+str(j) not in li:     #Check if exists from a list of original values that was already available at the start of the game
                    if Stochastic.check_line(puzzle, str(i)+str(j)):       #If the list isn't empty
                        error.append(str(i)+str(j))     #Fill the list with the location
                        #print('error')
                    #else:
                        #print('doesnt')
        return error

    def empty(puzzle, error):
        '''Empty the location with an error'''
        for value in error:
            i = int(value[0])   #Split the location to i and j
            j = int(value[1])
            puzzle[i][j] = 0    #Check the error to 0
    ################ END of Alec Kray's Code ###########################
#################### END of functions in STOCHASTIC ####################

#################### Pull Puzzle from File ####################
#################### Robyn's Code ####################
def pull_puzzle(difficulty):
    '''receives input as easy, medium, or hard then returns a random puzzle based off that'''
    #create the file name by combining the input string and the .txt extension
    file_name = difficulty + '.txt'
    #using file name, choose which file to open and create an empty list for each line
    file = open(file_name)
    #add each line to a list
    file_list = file.readlines()
    #return a random puzzle from the list, stripping the /n off
    return file_list[random.randrange(0,len(file_list))].strip()

def create_puzzle(puzzle):
    '''receives a string input of 81 numerical characters and returns a 9x9 list of lists'''
    multi_list = []
    #for every 9th number, starting from 0 and going to 72
    for a in range(0,81,9):
        #reset single list
        single_list = []
        #add each grouping of nine (0-9, 10-18, etc.) to the single list
        for b in puzzle[a:a+9]:
            single_list.append(int(b))
        #add the single list to the big list
        multi_list.append(single_list)
    return multi_list
#################### END of Robyn's Code ####################
#################### END of Pull Puzzle from File ####################

#################### Rediet's Code ####################
# Import graphics by assgining it to g
import graphics as g
# Import time 
import time 

# Create a function where text will be.
def text(x1, y1, win, word):
# Put in points, backgrounds, and word for the diffrent places.
    txt = g.Text(g.Point(x1, y1), word)
    txt.setFill(g.color_rgb(255,255,255))
    txt.draw(win)
    return txt

# Create a function for the rectangle.
def Rectangle(x1, y1, l, w, win):
# Put in points and add width and length to the varialble, and add the backgrounds.
    rect = g.Rectangle(g.Point(x1,y1), g.Point(x1+w, y1+l))
# Assgin color for line rectangle.
    rect.setOutline(g.color_rgb(186,206,195))
# Assgin the thickness of the triangle.
    rect.setWidth(2)
# Fill a color for the rectangle.
    rect.setFill(g.color_rgb(49,90,74))
    rect.draw(win)
    return rect

# Creat a function that will allow users to click
def mouse_selection(win):                                         #define the function for gui mouse selection
    ''' This function uses recursion'''     ################### Recursion ################################
    # Assgine points and coordinate for where the box might be      #get the cordinates for the rectangle for easy mode 
    p1 = win.getMouse()   
    p1x = p1.getX()
    p1y = p1.getY()
    print(p1x, p1y)
# Create an if statement to print diffrent puzzle depending on the diffculty    
    if 200> p1x > 150 and 130 > p1y > 90:   # The x positions of the easy box
        puzzle = create_puzzle(pull_puzzle('easy'))
        return puzzle
# Difrent else statements for diffrent puzzle depending on the diffculty    
    elif 285> p1x > 215 and 130 > p1y > 90:
        puzzle = create_puzzle(pull_puzzle('medium'))
        print('check')
        return puzzle
# Another elif statement for diffrent puzzle depending on the diffculty    
    elif 350 > p1x > 300 and 130 > p1y > 90:
        puzzle = create_puzzle(pull_puzzle('hard'))
        print('check')
        return puzzle
    else:
        return mouse_selection(win) 
    
# Create a main function where it call all the other function.
def main():
# Assgin a title for the graphics "sudoku solver"
    win = g.GraphWin("Sudoku Solver", 500,500)
# Assgine a color for the background
    win.setBackground(g.color_rgb(49,90,74)) #customizing colours no more than 225

# Create a for loop to repeat the first rectangle 9 times  
    for i in range(9):
# Create another for loop and adjust the points
        for j in range(9):
            Rectangle(20+(i*20), 160+(j*20), 20, 20, win)
            print()
    for i in range(9):
# Create another for loop and adjust the points
        for j in range(9):
            Rectangle(300+(i*20), 160+(j*20), 20, 20, win)
            print()

#################### END of Rediet's Code ####################
#################### Ruth's Code ####################
# # Call the rectangle function
    Rectangle(150, 90, 40, 50, win)
# Call the text function and assgin a word.
    text(175, 110, win, "Easy")
# Call the rectangle function          
    Rectangle(300, 90, 40, 50, win)
# Call the text function and assgin a word.
    text(325, 110, win, "Hard")
# Call the rectangle function 
    Rectangle(80, 30, 40, 70, win)
# Call the text function and assgin a word.
    back_timer = text(115, 50, win, "Timer")
# Call the rectangle function 
    Rectangle(350, 30, 40, 70, win)
# Call the text function and assgin a word.
    stoc_timer = text(385, 50, win, "Timer")
# Call the rectangle function
    Rectangle(215, 90, 40, 70, win)
# Call the text function and assgin a word.
    text(250, 110, win, "Medium")
# Call the rectangle function 
    Rectangle(45, 380, 50, 120, win)
# Call the text function and assgin a word.
    text(105, 405, win, "Backtracking")
# Call the rectangle function 
    Rectangle(340, 380, 50, 90, win)
# Call the text function and assgin a word.
    text(385, 405, win, "Stochastic")


    #################### End of Ruth's Code #################### 
    #################### Rediet's Code ####################   

    ori_puzzle = []     #Store puzzle value for stochastic algorithm
    for i in mouse_selection(win):
        ori_puzzle.append(tuple(i))     #Store value into tuples
    
    puzzle = []     #Puzzle for backtracking

    for i in ori_puzzle:
        puzzle.append(list(i))      #Change tuple to list and store it into puzzle

    print(puzzle)

    back_time = time.time()             #Back tracking time tracker
    
    back_timer.setText(str(round(time.time()-back_time,2)))


    back_sudo_text = [[],[],[],[],[],[],[],[],[]]
    #Backtracking
    for i in range(9):
    # Create another for loop and adjust the points
        for j in range(9):
            #Check if the puzzle is 0 if not call the text function
            if puzzle[i][j] != 0:
                #Adjust the points for the text in the first sudoku box
                back_sudo_text[i].append(text((j*20 + 30),(i*20 + 170), win, puzzle[i][j]))
            else:
                back_sudo_text[i].append(None)

    stoc_sudo_text = [[],[],[],[],[],[],[],[],[]]
    #Stochastic
    for i in range(9):
    # Create another for loop and adjust the points
        for j in range(9):
            #Check if the puzzle is 0 if not call the text function
            if puzzle[i][j] != 0:
                #Adjust the points for the text in the second sudoku box
                stoc_sudo_text[i].append(text((j*20 + 310),(i*20 + 170), win, puzzle[i][j]))
            else:
                stoc_sudo_text[i].append(None)

    #################### BACKTRACKING ####################
    #################### Robyn's Code ####################
    empty_spots = Backtracking.zero_finder(puzzle)
    #current defines what coordinate is being used
    current = 0
    error = 1
    while error != 0 and (time.time() - back_time) < 60:    
        #set current row and column values based off current and the empty spot list
        row = int(empty_spots[current][0])
        col = int(empty_spots[current][1])
        
        
        #print statements used for debugging
        #displays current, row, col, the current empty spot and the puzzle so far
        #print('current', current)
        #print('coordinate', str(row)+str(col))
        #print('pre', empty_spots[current], puzzle)
        
        
        # Remove the text to update it
        if back_sudo_text[row][col] != None:
            back_sudo_text[row][col].setText('')

        #increment current tile by one then set value equal to the current tile
        puzzle[row][col] += 1
        value = puzzle[row][col]
                    
        #Write the text into Graphics            
        back_sudo_text[row][col] = text((col*20 + 30),(row*20 + 170), win, puzzle[row][col])  
        
        #print statements used for debugging
        #displays current empty tile and puzzle post tile increment
        #print('post', empty_spots[current], puzzle)
        
        #if the current value is greater than the puzzle allows, reset the tile and step back
        if value < 10:
            #if the current tile has no errors, increment current
            if Backtracking.check_line(puzzle, empty_spots[current], value) == 0:
                #print('no errors found. next tile')

                #print statements used for debugging
                #displays current before increment
                #print('pre current', current)

                back_timer.setText(str(round(time.time()-back_time,2)))  # Track time
                
                #increment current
                current += 1

                #print statements used for debugging
                #displays current after increment
                #print('post current', current)

        else:
            #print('value greater than four. attempting reset')
            
            #print statements used for debugging
            #displays current and tile value pre-adjustment
            #print('pre current', current)
            #print('pre value', puzzle[row][col])
            
            #resets tile value to zero and decrements current in order to move to previous empty space
            puzzle[row][col] = 0

            if back_sudo_text[row][col] != None:
                back_sudo_text[row][col].setText('')  #Move the text away from the screen
                
            current = current - 1
            
            #print statements used for debugging
            #displays current and tile value post-adjustment
            #print('post current', current)
            #print('post value', puzzle[row][col])
        
        #if the puzzle is full, check every tile in empty_spots to make sure there are no errors
        if Backtracking.full(puzzle) != 0:
            #count the amount of errors, set error = 0 initially
            error = 0
            #for every tile, add 1 for every time an error occurs
            for a in empty_spots:
                error += Backtracking.check_line(puzzle, a, puzzle[int(a[0])][int(a[1])])
        
        #print('END OF WHILE LOOP')
    print("SOLVED PUZZLE")
    for i in range(9):
        print(puzzle[i])
    #################### END of Robyn's Code ####################

    if (time.time() - back_time) > 60:
        failed = Rectangle(65, 220, 40, 70, win)
        failed.setOutline(g.color_rgb(0,0,0))
        failed.setFill(g.color_rgb(255, 0, 0))
        text(100, 240, win,'FAILED!')

    back_time = time.time() - back_time
    print(back_time)

    stoc_time = time.time()
    stoc_timer.setText(str(round(time.time()-stoc_time,2)))     # Update the timer
    #################### END of BACKTRACKING ####################
    #################### STOCHASTIC ####################
    puzzle= []

    for i in ori_puzzle:
        puzzle.append(list(i))

    print(puzzle)
    stoc_time = time.time()

    #print(puzzle)
    present_dict = Stochastic.count_grid(puzzle) #(value,number)   #Dictionary of current values

    missing_list = Stochastic.create_missing(present_dict) #List of missing values

    #print('Missing:',len(missing_list)) 

    original_list = Stochastic.original(puzzle)    #List of original values
    Stochastic.filler(puzzle, missing_list)        #Fill the puzzle

    for i in range(9):
    # Create another for loop and adjust the points
        for j in range(9):
            #Check if the puzzle is 0 if not call the text function
            if puzzle[i][j] != 0:
                #Adjust the points for the text in the second sudoku box
                stoc_sudo_text[i][j] = (text((j*20 + 310),(i*20 + 170), win, puzzle[i][j]))


    #print('Original List:',original_list)

    err = Stochastic.checker(puzzle, original_list)    #Check for error

    ################ END of Alec Kray's Code ###########################

    ################ Abemelech's Code ###########################
    ###Faster Algorithm###
    orginal_err = err       #Store the orriginal locaiton of the errors

    while len(err) != 0 and (time.time() - stoc_time) < 60:    #Check if there is an error and loop until there isn't any error and have a time limit

        init_list = tuple(err)  #Save the original list of errors to compare at the end if the shuffling didn't work

        for local in err:       #Loop through the errors
            #print(local)
            for i in range(9):      #Loop through the entire grid
                for j in range(9):
                    if str(i)+str(j) != local and str(i)+str(j) not in original_list:       #Check position is a changable number (or grid)
                        num_error = len(err)  #Store the number of error

                        stoc_timer.setText(str(round(time.time()-stoc_time,2))) # Update the timer

                        stoc_sudo_text[int(local[0])][int(local[1])].setText('')         #Move both text of the two numbers from the screen
                        stoc_sudo_text[i][j].setText('')                                  

                        puzzle[int(local[0])][int(local[1])], puzzle[i][j] = puzzle[i][j], puzzle[int(local[0])][int(local[1])] # Switch the two numbers in the grid,  the error cell with a random cell


                        stoc_sudo_text[int(local[0])][int(local[1])] = text((int(local[1])*20 + 310),(int(local[0])*20 + 170), win, puzzle[int(local[0])][int(local[1])])   #Write the text for both the two numbers being switched
                        stoc_sudo_text[i][j] = text((j*20 + 310),(i*20 + 170), win, puzzle[i][j])

                        if len(Stochastic.checker(puzzle, original_list)) < num_error:     #Check if the number of error decreased
                            err = Stochastic.checker(puzzle, original_list)                #If the number of error decreased update the error list
                            #print('worked',err)
                        else:                               #If not switch back the two numbers in the grid
                            
                            stoc_sudo_text[int(local[0])][int(local[1])].setText('')        #Move both text of the two numbers from the screen
                            stoc_sudo_text[i][j].setText('')
                            
                            puzzle[i][j], puzzle[int(local[0])][int(local[1])] = puzzle[int(local[0])][int(local[1])], puzzle[i][j]
                            stoc_sudo_text[int(local[0])][int(local[1])] = text((int(local[1])*20 + 310),(int(local[0])*20 + 170), win, puzzle[int(local[0])][int(local[1])])   #Write the text for both the two numbers being switched
                            stoc_sudo_text[i][j] = text((j*20 + 310),(i*20 + 170), win, puzzle[i][j])
        
        #After running through all the error's if the the number didn't change, then start over
        if init_list == tuple(err):
            Stochastic.empty(puzzle, orginal_err)      #Re empty the grid
            present_dict = Stochastic.count_grid(puzzle)   # Check the present numbers
            missing_list = Stochastic.create_missing(present_dict) # Store the missing numbers
            Stochastic.filler(puzzle, missing_list)    #Fill back the grid
            #print('check', puzzle)
            err = Stochastic.checker(puzzle, original_list)   #Check for errors
            #print(err)         
        #print(len(err))

    #print(puzzle)

    ### END OF ALGORITHM ###
    ################ END of Abemelech's Code ###########################

    ################ Alec Kray's Code ###########################
    ### Initial Algorithm ###
    '''counter = 0
    while len(err) != 0:                    #If there is an error since the len will be greater than 0
        empty(puzzle, err)                  #Empty the error grid
        present_dict = count_grid(puzzle)   #Do it all over again except for storing the original value since the original value doesn't change
        missing_list = create_missing(present_dict)
        filler(puzzle, missing_list)
        err = checker(puzzle, original_list)
        counter += 1
        #print(counter, end =" ")'''

    ### END OF ALGORITHM ###
    #################### END of ALec Kray's Code ####################
    #################### END of STOCHASTIC ####################
    if (time.time() - stoc_time) > 60:
        failed = Rectangle(350, 220, 40, 70, win)
        failed.setOutline(g.color_rgb(0,0,0))
        failed.setFill(g.color_rgb(255, 0, 0))
        text(385, 240, win,'FAILED!')

    #If no errors print the solved puzzle
    print('Solved Puzzle') 
    for hor in puzzle:
        print(hor)   

    stoc_time = time.time() - stoc_time
    print(stoc_time)

#     time.sleep(1)
#     number6 = text(120,100, win,"6")
#     time.sleep(1)
#     number6.move(80,100)
#     time.sleep(1)
   

#  Get mouse
    win.getMouse()
    win.close()
# Call the main function  
main()
################ End of Rediet's Code ###########################   
```
