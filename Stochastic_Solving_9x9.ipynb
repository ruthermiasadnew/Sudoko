```python
###Abemelech's Code###
import random

#4x4 SUDOKU SOLVER


grid = [[3, 0, 4, 0], #original 3 0 4 0
        [0, 1, 0, 2], #orignal 0 1 0 2
        [0, 4, 0, 3], #original 0 4 0 3
        [2, 0, 1, 0]] #original 2 0 1 0

def check_line(grid, location):
    '''This function accepts two inputs of a grid of lists of lists (4x4) and a string of the location we
    want to check. which can be like '00' or '32', where the first location is the first of the string    
    '''
    error = []      #An empty list that will contain all the errors on that location
    
    #mover = [[1,2,3],[10,20,30]]
    #for i in range(4):
    #    for location in mover:         #Loop through the horizontal and vertial potion of the 00 position
    #        for value in location:             
    #            print(starter+value)   #These are the location we need to check for the 00 position
    #    mover[0][i] *= -1
    #    print(mover, starter)
    

    x = int(location[1])    #Save the string into x and y values
    y = int(location[0])

    if grid[y][x] != 0:         #check if the grid is different from 0, so it will only check filled grids
        #horizonal check
        for hor in range(9):    #loop through the 0 to 3
            if hor != x:        #Avoid checking the current position so if it is in '00' don't want to check '00' again
                if grid[y][hor] == grid[y][x]:          #Check if the location passed is equal to every value in the horizontal
                    error.append(str(y)+str(hor))       #If so it is an error  

        #vertical check
        for ver in range(9):    #loop through 0 to 3
            if ver != y:        #Avoid checking the current position so if it is in '00' don't want to check '00' again (in the y axis)
                if grid[ver][x] == grid[y][x]:      #Check if the location passed is equal for very value in vertical
                    error.append(str(ver)+str(x))   #Add to error

        #Check per box
        if x//3 == 0 and y//3 == 0:         #Check for box 1, by divide the values of the x and y values we can guess where in the box it is found
            for hor in range(3):        #For box 1 only values 0 and 1 are needed
                for ver in range(3):
                    if hor != x and ver != y:       #Check if it is equal to the passed value
                        if grid[ver][hor] == grid[y][x]:        #Check if the passed location's value is equal to all the value in the box
                            error.append(str(ver)+str(hor))     #If so add it into error

        elif x//3 == 1 and y//3 == 0:      #Check for box 2
            for hor in range(3,6):
                for ver in range(3):    #For box 2: 0,1 for the vertical and horizional value 2,3 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))

        elif x//3 == 2 and y//3 == 0:      #Check for box 3
            for hor in range(6,9):
                for ver in range(3):    #For box 3: 0,1 for the vertical and horizional value 2,3 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))

        elif x//3 == 0 and y//3 == 1:      #Check for box 4
            for hor in range(3):
                for ver in range(3,6):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))
        elif x//3 == 1 and y//3 == 1:
            for hor in range(3,6):      #Check for box 5
                for ver in range(3,6):  #For box 4: 2,3 is what needed for hor and vert and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))

        elif x//3 == 1 and y//3 == 1:
            for hor in range(6,9):      #Check for box 6
                for ver in range(3,6):  #For box 4: 2,3 is what needed for hor and vert and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))
        
        elif x//3 == 0 and y//3 == 2:      #Check for box 4
            for hor in range(3):
                for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))

        elif x//3 == 1 and y//3 == 2:      #Check for box 4
            for hor in range(3,6):
                for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))

        elif x//3 == 2 and y//3 == 2:      #Check for box 4
            for hor in range(6,9):
                for ver in range(6,9):  #For box 3: 0,1 for the horizonal and vertical value 2,3 is what needed and do the same by checking
                    if hor != x and ver != y:
                        if grid[ver][hor] == grid[y][x]:
                            error.append(str(ver)+str(hor))
        
        return error




#print(grid)

def count_grid(grid):
    ''' This counts the number of values
    Ex: there is 2 0's and 3 4's and so forth and so one.
    It passed the grid to check for how much of a single value is available 
    then it stored in a dictionary
    '''
    counter = {}        #Empty dictionary
    for hor in grid:    #Loop through all the values of the grid  
        for value in hor:
            if value != 0:
                if value in counter:
                    counter[value] += 1     #Add one if the value exists
                else:
                    counter[value] = 1      #Create one if the value doesn't exist
    return counter

def create_missing(counter):
    '''The function creates a list of missing values that could
    later be filled. It accepts a dictionary of the current available values
    '''
    possible_value = []         #Empty list to store the values
    for key,value in counter.items():   #loop through the dictionary
        for _ in range(9-value):        #Loop through the number of times it is empty and add the value to the list
            possible_value.append(key) 
    return possible_value

```

```python
###Alec Kray's Code###

def original(puzzle):
    '''Create List of original puzzle values'''
    org = []        #List to store
    for i in range(9):      #Loop through the grid to store the original puzzle location
        for j in range(9):
            if puzzle[i][j] != 0:
                  org.append(str(i)+str(j))
    return org

def filler(puzzle,list):
    '''Fill the puzzle randomly based on a list'''
    for horizontal in puzzle:   #Loop through the puzzle
        for i in range(9):      
            #print(horizontal[i])
            if horizontal[i] == 0:  #Check if puzzle value is zero
                horizontal[i] = random.choice(list)     #Fill randomly from list       
        
            
            



def checker(puzzle,li):
    ''' Check the entire puzzle if there is an error'''
    error = []      #Empty list
    for i in range(9):      #Loop through i and j 0 to 3
        for j in range(9):
            if str(i)+str(j) not in li:     #Check if exists from a list of original values that was already available at the start of the game
                if check_line(puzzle, str(i)+str(j)):       #If the list isn't empty
                    error.append(str(i)+str(j))     #Fill the list with the location
                    #print('error')
                #else:
                    #print('doesnt')
    return error

def empty(puzzle, error):
    '''Empty the location with an error'''
    for value in error:
        i = int(value[0])   #Split the location to i and j
        j = int(value[1])
        puzzle[i][j] = 0    #Check the error to 0
            



def main():

    puzzle = [[0,2,0,5,0,1,0,9,0],    #Puzzle we are going to check with
            [8,0,0,2,0,3,0,0,6],
            [0,3,0,0,6,0,0,7,0],
            [0,0,1,0,0,5,6,0,0],
            [5,4,0,0,2,0,0,1,9],
            [0,0,2,0,4,0,7,0,0],
            [0,9,0,0,3,0,0,8,1],
            [2,0,0,8,0,4,9,0,7],
            [0,1,8,9,0,7,0,6,0]]

    print(puzzle)
    present_dict = count_grid(puzzle) #(value,number)   #Dictionary of current values

    missing_list = create_missing(present_dict) #List of missing values

    print('Missing:',missing_list) 

    original_list = original(puzzle)    #List of original values
    filler(puzzle, missing_list)        #Fill the puzzle

    print('Original List:',original_list)

    err = checker(puzzle, original_list)    #Check for error

    while len(err) != 0:                    #If there is an error since the len will be greater than 0
        empty(puzzle, err)                  #Empty the error grid
        present_dict = count_grid(puzzle)   #Do it all over again except for storing the original value since the original value doesn't change
        missing_list = create_missing(present_dict)
        filler(puzzle, missing_list)
        err = checker(puzzle, original_list)
        
    print(puzzle)

main()
```
